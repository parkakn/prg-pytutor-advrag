#### 1) Assignments

##### 1.a) docx to pdf

```python
import subprocess

def convert_docx_to_pdf(docx_path):
    # Get the directory of the original DOCX file
    output_dir = os.path.dirname(docx_path)
    
    # Convert DOCX to PDF and save in the same folder
    subprocess.run(['libreoffice', '--headless', '--convert-to', 'pdf', '--outdir', output_dir, docx_path])
```

```python
files_assignments = glob.glob("GSU/Assignments/*.docx")  
```

```python
for i in files_assignments:
    convert_docx_to_pdf(i)
    print(f"Converted {i} to PDF")
```

##### 1.b) pdf to markdown

```python
from rag_tools.document_parsing import documentparser
```

```python
# Load environment variables from .env file
load_dotenv()

nest_asyncio.apply()

openai_api_key = os.environ.get("OPENAI_API_KEY")
llama_cloud_api_key = os.getenv("LLAMA_CLOUD_API_KEY")
```

```python
files_assignments = glob.glob("GSU/Assignments/*.pdf")  
```

```python
files_assignments
```

```python
model = "openai-gpt4o"
```

```python
md_assignments = documentparser.document_to_markdown(files_assignments,"assignments",model)
```

```python
print(md_assignments[0][0].text)
```

```python
for i in range(len(md_assignments)):    
    txt_file_name = os.path.basename(files_assignments[i]).replace('.pdf', '.txt')
    
    # Save the course summary to a text file
    with open(f'''GSU/Assignments/{txt_file_name}''', "w") as file:
        file.write(md_assignments[i][0].text)
```
